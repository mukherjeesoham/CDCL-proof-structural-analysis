## Classiyfing SAT instances
Here's the question we want to answer. Given several families SAT instances and their solving times on a specific CDCL solver, can we
- build a classifier that assigns a new SAT instance to one of the families? 
- build a classifier that classifies a SAT instance within a particular family as easy or hard?

Here's a roadmap. Visualize the data using t-SNE [[1]](https://distill.pub/2016/misread-tsne/). If it looks 'good', do a PCA and project the data to a lower dimensional subspace. Then run a random forest or a KNN classifier. Test the accuracy of your classifier. If it's really bad, one still has a few options. One can increase the number of neighbours in their KNN classifier; be aware however, that you might run into overfitting problems if you choose $k$ to be really high. One could also change the number of eigenvalues to keep. Finally, look for new features one could compute for the graph; for example, small world networks.

```python
datasets = {
    "verification" : (0, pd.read_pickle("../datasets/pickle/all_verification.names.pkl")),
           "agile" : (1, pd.read_pickle("../datasets/pickle/all_agile.names.pkl")),
          "crypto" : (2, pd.read_pickle("../datasets/pickle/all_crypto.names.pkl")),
         "crafted" : (3, pd.read_pickle("../datasets/pickle/all_crafted.names.pkl")),
          "random" : (4, pd.read_pickle("../datasets/pickle/all_random.names.pkl"))
}

hardness = {
    "easy" : np.log2(60),
    "hard" : np.inf
}
```

Build the whole dataset with labels

```python
def labelhardness(solving_time, hardness):
    bins = []
    for key in hardness:
        bins.append(hardness[key])
    bins = np.array(bins)
    return np.digitize(solving_time, bins)
     
def labelfamily(dataset, family):
    A = dataset
    B = np.repeat(family, np.shape(dataset)[0]).reshape(-1, 1)
    return np.hstack((A,B))

def label(name, datasets, hardness):
    family = datasets[name][0]
    dataset = datasets[name][1].to_numpy()
    dataset[:,-1] = labelhardness(dataset[:,-1], hardness)
    dataset = labelfamily(dataset,family)
    return dataset
 
def merge(datasets, hardness):
    merged = np.zeros((1,513))
    for key in datasets:
        merged = np.vstack((merged, label(key, datasets, hardness)))
    return merged[1:, :]
```

Separate data into features and predictors, and scale the data.

```python
XY = merge(datasets, hardness)
X  = StandardScaler().fit_transform(XY[:, 0:-2])
Y  = XY[:, -2:]
```

```python
def colour(feature):
    colours = {0 : "r", 1 : "b", 2 : "g", 3 : "c", 4 : "k"}
    return list(map(lambda x : colours[x], feature))
```

```python
def tSNE(features, colors, iterations):
    perplexity = [20, 30, 40, 50]
    fig = plt.figure(figsize=(20,4))
    for (index, perp) in enumerate(perplexity):
        features_embedded = TSNE(n_components=2, perplexity=perp, n_iter=iterations).fit_transform(features)
        plt.subplot(1, 4, index+1)
        plt.scatter(features_embedded[:,0], features_embedded[:,1], color=colors)
        plt.title(perp)

def project(features, dimension):
    print(PCA(n_components=dimension).fit(features).singular_values_)
    return PCA(n_components=dimension).fit_transform(features)
```

Throw everything at it. All that sklearn has to offer.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

names = ["Nearest Neighbors", "Linear SVM", "RBF SVM",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes"]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025), # SVC are sensitive. Type of kernel is important. 
    SVC(gamma=2, C=1),
    DecisionTreeClassifier(max_depth=5), # Can overfit
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1), # Good at avoiding overfitting > esmiators > robust
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(), # avoids overfitting (XGboost); sequential instead of parallel. Change the number of iterations.
    GaussianNB()
]
```

```python
for name, clf in zip(names, classifiers):
    scores = cross_val_score(clf, X, Y[:,-1], cv=10)
    print(name + ": Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
